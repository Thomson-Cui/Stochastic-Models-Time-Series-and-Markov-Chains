$\textbf{1.}$ For a Markov chain that satisfies the convergence theorem and so

$
p^n(x, y) \to \pi(y)
$

a natural question is how fast this convergence is. We can quantify this by finding some function of $n$, $f(n)$, with the property that $f(n) \to 0$ as $n \to \infty$ such that (at least for large enough $n$)

$
\sum_{y} |p^n(x, y) - \pi(y)| \leq f(n)
$

We can investigate this using the technique from the proof of the convergence theorem (theorem 1.19) of considering two independent copies of the Markov chain defined by $p$, $X_n$ and $Y_n$, where the initial distribution of $X_1$ is arbitrary but the initial distribution of $Y_n$ is $Y_n \sim \pi(y)$.

Consider a Markov transition matrix $p$ on the state space $\{0,1\}$:

$$
\begin{array}{c|cc}
 & 0 & 1 \\
\hline
0 & .6 & .4 \\
1 & .4 & .6 \\
\end{array}
$$

$\textbf{(a)}$ Suppose $X_n$ and $Y_n$ are independent Markov chains that both transition according to $p$. Let $T = \min\{n \geq 1: X_n = Y_n\}$. Show that regardless of the initial distributions of $X_n$ and $Y_n$

$
P(T>n) \leq (1 - .4^2)^n
$

(Hint: Suppose by time $n-1$ we have not yet observed $X_{n-1} = Y_{n-1}$. What can we say about $P(X_n = Y_n)$, that is, they agree after the next step?)

$\pi P=\pi$

$\pi=(0.5,0.5)$

$\begin{aligned}
&P(X_n = Y_n \mid X_{n-1} \neq Y_{n-1})\\
&= P(X_n = 1 \mid X_{n-1} = 0) \cdot P(Y_n = 1 \mid Y_{n-1} = 1)\\ 
&+ P(X_n = 0 \mid X_{n-1} = 1) \cdot P(Y_n = 0 \mid Y_{n-1} = 0)\\
&= (0.4)(0.6) + (0.4)(0.6) = 0.24 + 0.24 = 0.48
\end{aligned}$

$P(X_n \neq Y_n \mid X_{n-1} \neq Y_{n-1}) = 1 - 0.48 = 0.52.$

$\begin{aligned}
&P(T > n)\\
& \leq P(X_n \neq Y_n \mid X_{n-1} \neq Y_{n-1}) \cdot P(X_{n-1} \neq Y_{n-1} \mid X_{n-2} \neq Y_{n-2}) \cdots P(X_1 \neq Y_1)\\
& = (0.52)^n
\end{aligned}$

$P(T > n) \leq (0.52)^n \leq (0.84)^n=(1 - 0.4^2)^n$

Therefore, $P(T > n)\leq (1 - 0.4^2)^n$

$\textbf{(b)}$ Now suppose that the initial distribution of $X_n$ is arbitrary but the initial distribution of $Y_n$ is $Y_0 \sim \pi(y)$. Show that

$
\sum_{y}|p^{n}(x,y)-\pi(y)|\le2(1-.4^{2})^{n}
$

(Hint: Use the techniques from steps 3 and 4 of the proof of theorem 1.19)

$P^n(x,y)\to\pi(y)$ as $n\to\infty$

- $ X_n $ has an arbitrary initial distribution.
- $ Y_n $ has an initial distribution $ Y_0 \sim \pi $, where $ \pi $ is the stationary distribution of the Markov chain.

For any state $y$, we can write:
- $p^n(x,y)=P(X_n=y)$
- $\pi(y)=P(Y_n=y)$

    Thus, $|p^{n}(x,y)-\pi(y)|=|P(X_n=y)-P(Y_n=y)|$

$ X_n = Y_n $ for all $ n \geq T $. 
- For $ n \geq T $, $ P(X_n = y) = P(Y_n = y) $.
- For $ n < T $, $ X_n $ and $ Y_n $ are independent, and their distributions may differ. Thus

$|P(X_n = y) - P(Y_n = y)| \leq P(T > n).$

$\sum_{y} |p^n(x, y) - \pi(y)| \leq 2P(T > n).$

The factor of 2 arises because the total variation distance between two distributions is bounded by twice the probability that the coupled chains have not yet met.

Since $P(T > n) \leq (1 - 0.4^2)^n.$, we can get:

$\sum_{y} |p^n(x, y) - \pi(y)| \leq 2(1 - 0.4^2)^n.$


$\textbf{(c)}$ Now we will generalize this idea. Suppose a transition matrix $p$ on a finite state space has the property that $p(x,y)>0$ for all $x, y$. Show that there exist constants $C_1$ and $C_2$ such that

$

$

Since $ p(x, y) > 0 $ for all $ x, y $, there exists a constant $ \delta > 0 $ such that:

$p(x, y) \geq \delta \quad \text{for all } x, y.$

This implies that at each step, the probability that $ X_n = Y_n $ is at least $ \delta^2 $, because:

$P(X_n = Y_n \mid X_{n-1} = x, Y_{n-1} = y) = p(x, y) p(y, y) \geq \delta^2.$

Thus, the probability that $ X_n \neq Y_n $ at each step is at most $ 1 - \delta^2 $. Using this, we can bound $ P(T > n) $ as:

$P(T > n) \leq (1 - \delta^2)^n.$

So, 
$\sum_{y} |p^n(x, y) - \pi(y)| \leq 2(1 - \delta^2)^n.$


Therefore,  exists $ C_1 = 2 $ and $ C_2 = -\log(1 - \delta^2) $ to make $\sum_{y} |p^n(x, y) - \pi(y)| \leq C_1 e^{-C_2 n}.$
